## Description of Columns

* SK_ID_PREV

前の応募時のIDと思われる。Home Creditは、最初に小口のPOS CASHを貸しつけてそれが返済できる人に大口融資をするスタイル。

* SK_ID_CURR

現在の融資でのID

* MONTH_BALANCE

おそらく現在の応募に対する月情報

* CNT_INSTALMENT

* CNT_INSTALMENT_FUTURE

* NAME_CONTRACT_STATUS

* SK_DPD

* SK_DPD_DEF

## 単純にIDは通し番号の可能性

### SK_ID_PREVとSK_ID_CURR

#### UniqueなSK_ID_PREVはいくつ？

```python
pos.SK_ID_PREV.nunique()
```

#### Groupbyして平均個数をとる

```python
grp_prev = pos.groupby('SK_ID_PREV')
grp_prev['SK_ID_PREV'].count().mean()
```

#### SK_ID_PREVの個数の分布

```python
%matplotlib inline
grp_prev['SK_ID_PREV'].count().plot(kind='hist', bins=96)
plt.savefig('../figure/POS_CASH_ID_PREV_count_hist.png')
```

#### 同一SK_ID_PREVの個数は何通り?

```python
grp_prev['SK_ID_PREV'].count().nunique()
```

* 1-96まで全部存在するらしい

#### 同一SK_ID_PREVのデータを眺めてみる

```python
pos.sort_values(by='SK_ID_PREV', ascending=True)
```

## カスタム特徴

### 初期の申請のときの分割から変更をしたか

```python
pos.groupby('SK_ID_CURR').agg({
    'SK_ID_CURR': 'size',
    'SK_ID_PREV': 'nunique'
}, columns={
    'SK_ID_CURR': 'POS_COUNT',
    'SK_ID_PREV': 'POS_N_CONTRACT'
})
```

```python
pos.groupby('SK_ID_PREV').agg({'SK_ID_CURR': 'nunique'})
```
